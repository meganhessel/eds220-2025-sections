## Subsetting 

```python
prey.shape
```

```python
# Select only the preserved speciimes from prey 
preserved = prey [ prey['basisOfRecord']== 'PRESERVED_SPECIMEN' ]

print(preserved.shape)
```

```python
# Select obserbations beteen 1992 and 2000 
sub_yr = prey [ prey['year'].between(1992,2000) ]

sub_yr2 = prey [ (prey['year'] >= 1992) & (prey['year'] <= 2000)]

print(sub_yr.shape)
print(sub_yr2.shape)
```

```python
# Select all rows for columns species and eventData 
prey_sub = prey [ ['species', 'eventDate'] ]

print(prey_sub.shape)
```

```python
# Select observations that are greater than 37 lat and less than 39 lat 
prey_lat = prey[ (prey['lat']>37) & (prey['lat'] < 39) ]

print(prey_lat.shape)
```
